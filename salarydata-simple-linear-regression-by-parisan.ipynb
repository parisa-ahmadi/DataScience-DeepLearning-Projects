```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = pd.read_csv("../input/salary-data-simple-linear-regression/Salary_Data.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
df.value_counts()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(10,8), dpi=80)

plt.scatter(df["YearsExperience"],df["Salary"]) 

plt.title("YearsExperience vs Salary", color="green", size=30)

plt.xlabel("YearsExperienceYearsExperience", c="blue", size=25)

plt.ylabel("Salary", c="red", size=25)

plt.grid()

plt.show()
```

```python
x = df["YearsExperience"].values.reshape(-1,1)
y = df["Salary"].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
#Intercept
print(regressor.intercept_)

#slope
print(regressor.coef_)
```

```python
plt.scatter(x_test, y_test, color = "gray")

plt.plot(x_test, y_pred, color = "red")

plt.show()
```

```python
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error: ", metrics.mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error: ", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("R2 Score: ", metrics.r2_score(y_test, y_pred))
```

```python
plt.scatter(x_test, y_test, color = "gray")

plt.scatter(x_test, y_pred, color = "red")

plt.show()
```

```python
plt.figure(figsize=(10,8), dpi = 80)

plt.scatter(y_test, y_pred)

plt.grid()
```
